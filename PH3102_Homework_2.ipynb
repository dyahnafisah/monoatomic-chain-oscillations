```python
# importing the necessary packages

import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
import pandas as pd
from scipy import optimize
```

```python
# defining global variables

global N, k, m, a
N = 16
k = 1
m = 1
a = 1
```

```python
# create K matrix

K = np.zeros((N, N))
for i in range(N):
  K[i][(i-1)%N] = 1
  K[i][i%N] = -2
  K[i][(i+1)%N] = 1
```

```python
np.set_printoptions(linewidth=np.inf)
K
```

```python
# solve for eigenvalues and eigenvectors, then sort by the magnitude of the eigenvalues

eigenvalues, eigenvectors = LA.eig(np.array(K))
idx = eigenvalues.argsort()[::]   
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:,idx]
```

```python
# store the eigenvalues and eigenvectors into dataframe and display

df = pd.DataFrame()
df['Eigenvalues'] = [round(y, 3) for y in eigenvalues]

for i in range(N):
  y = []
  for j in range(N):
    y.append(round(eigenvectors[i][j], 3))
  df["u"+str(i)] = y
df
```

```python
# define sinusoidal function to compare with the result that we got

def sin_func(x, a, b, c):
    return a * np.sin(b * x + c)
```

```python
# plot without fitting

for i in range(N):
  y = df.iloc[i,1:]
  x = [n for n in range(N)]
  print("\n\nEigenvalue =", df.iloc[i,0], ", w =", round(np.sqrt(abs(df.iloc[i,0])), 3))
  plt.plot(x, y)
  plt.xlabel("n")
  plt.ylabel("amplitude")
  plt.show()
```

```python
# plot and fit with sinusoidal function

for i in range(N):
  y = df.iloc[i,1:]
  x = [n for n in range(N)]

  params, params_covariance = optimize.curve_fit(sin_func, x, y, p0 = [1, round(np.sqrt(abs(df.iloc[i,0])), 3), 0])
  location = np.arange(0, 15, 0.1)
  amplitude = params[0] * np.sin(params[1] * location + params[2])

  print("\n\nEigenvalue =", df.iloc[i,0], ", w =", round(np.sqrt(abs(df.iloc[i,0])), 3))
  print("Fitting function: %.2fsin(%.2ft + %.2f)" % (params[0], params[1], params[2]))
  plt.scatter(x, y)
  plt.plot(location, amplitude, color='r')
  plt.xlabel("n")
  plt.ylabel("amplitude")
  plt.show()
```

```python
# sort the omega to plot the dispersion relation

j = N-1
l = 0
eigenvalues_sorted = [0] * N
w = [0] * N
for i in range(N):
  if ((i%2) == 0):
    eigenvalues_sorted[j] = round(eigenvalues[i], 3)
    w[j] = round(np.sqrt(abs(eigenvalues[i])), 3)
    j += -1
  else:
    eigenvalues_sorted[l] = round(eigenvalues[i], 3)
    w[l] = round(np.sqrt(abs(eigenvalues[i])), 3)
    l += 1

df_w = pd.DataFrame()
df_w['Eigenvalue'] = eigenvalues_sorted
df_w['Frequency (w)'] = w
df_w
```

```python
# list the values of the wavevectors

p = [i for i in range(int(-1*N/2 + 1), int(N/2 + 1))]
q = [2*np.pi*i/(a*N) for i in p]
df_k = pd.DataFrame()
df_k['p'] = p
df_k['k'] = q
df_k
```

image.png

```python
# expected dispersion relation

def dispersion_relation(q):
  return np.sqrt(4*k/m)*abs(np.sin((q*a)/2))

w_dispersion_relation = []
for n in q:
  w_dispersion_relation.append(dispersion_relation(n))
```

```python
# plot both actual and expected values of w wrt q

plt.plot(q, w_dispersion_relation, color='r')
plt.scatter(q, w)
plt.ylabel('Frequency (w)')
plt.xlabel('q')
plt.title("Dispersion relation")
plt.legend(["expected w", "actual w"], loc ="lower right")
plt.show()
```
